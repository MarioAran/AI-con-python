# Covariaza y Correlación

```python
%matplotlib inline

import numpy as np
from pylab import *

datosX = [173,170,168,180,169,175,160,172,169,163,192,169,174,165]
datosY = [18.5,18.5,17.5,19.0,16.5,19.2,16.8,18.5,18.1,17.0,21.5,17.7,18.1,17.3]

scatter(datosX, datosY)

print("Covarianza")
print(np.cov(datosX, datosY, bias=True))
print("Correlación")
print(np.corrcoef(datosX, datosY))

```
