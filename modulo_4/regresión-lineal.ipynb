# Regresi√≥n Lineal

```python
import numpy as np
import matplotlib.pyplot as plt

# Datos
datosX = [0, 4, 8]
datosY = [5, 5, 8]

# Crea el gr√°fico de dispersi√≥n (scatter plot)
plt.scatter(datosX, datosY)

# Opcional: Personaliza el gr√°fico
plt.title("Mi Gr√°fico de Dispersi√≥n")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")

# Muestra el gr√°fico
plt.show()
```

La funci√≥n lineregress retorna varios valores que nos interesan

```python
from scipy import stats

slope, intercept, r_value, p_value, std_err = stats.linregress(datosX, datosY)
```

```python
print(slope)
print(intercept)
```

```python
r_value ** 2
```

Ahora usaremos la pendiente y la intersecci√≥n que retorn√≥ la regresi√≥n para dibujar valores predecidos vs los observados:

```python
# Retorna un ndarray (N-dimensional array) NumPy permite realizar operaciones 
# matem√°ticas en todos los elementos del arreglo de manera simult√°nea, 
# sin necesidad de usar bucles expl√≠citos. 
# Esto se conoce como vectorizaci√≥n y es la raz√≥n principal de su alta velocidad.

x = np.linspace(0,8,10) # Crea un arreglo unidimensional de n√∫meros con un espaciado uniforme dentro de un intervalo espec√≠fico.
y = slope * x + intercept

plt.scatter(datosX, datosY)
plt.plot(x, y, c='r')
plt.show()
```
