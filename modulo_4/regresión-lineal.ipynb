# Regresión Lineal

```python
import numpy as np
import matplotlib.pyplot as plt

# Datos
datosX = [173,170,168,180,169,175,160,172,169,163,192,169,174,165]
datosY = [18.5,18.5,17.5,19.0,16.5,19.2,16.8,18.5,18.1,17.0,21.5,17.7,18.1,17.3]

# Crea el gráfico de dispersión (scatter plot)
plt.scatter(datosX, datosY)

# Opcional: Personaliza el gráfico
plt.title("Mi Gráfico de Dispersión")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")

# Muestra el gráfico
plt.show()
```

La función lineregress retorna varios valores que nos interesan

```python
from scipy import stats

slope, intercept, r_value, p_value, std_err = stats.linregress(datosX, datosY)
```

```python
print(slope)
print(intercept)
```

```python
r_value ** 2
```

Ahora usaremos la pendiente y la intersección que retornó la regresión para dibujar valores predecidos vs los observados:

```python
# Retorna un ndarray (N-dimensional array) NumPy permite realizar operaciones 
# matemáticas en todos los elementos del arreglo de manera simultánea, 
# sin necesidad de usar bucles explícitos. 
# Esto se conoce como vectorización y es la razón principal de su alta velocidad.

x = np.linspace(150,200,300) # Crea un arreglo unidimensional de números con un espaciado uniforme dentro de un intervalo específico.
y = slope * x + intercept

plt.scatter(datosX, datosY)
plt.plot(x, y, c='r')
plt.show()
```
