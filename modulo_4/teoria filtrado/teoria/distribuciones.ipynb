# Ejemplos de distribuciones de datos

## Distribución Uniforme

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

#El rango va de 1 a 6 son las caras de un dado
values = np.random.uniform(1, 6, 10000)
plt.hist(values, 6)
plt.show()
```

## Normal / Gaussiana

```python
from scipy.stats import norm
import matplotlib.pyplot as plt

#Return evenly spaced values within a given interval.
x = np.arange(-3, 3, 0.001)
print(x)
#plot(x, y)
plt.plot(x, norm.pdf(x))
```

Generamos números aleatorios con distribución normal. "mu" es la media, "sigma" la desviación estandar.

```python
import numpy as np
import matplotlib.pyplot as plt

mu = 5.0
sigma = 2.0
values = np.random.normal(mu, sigma, 10000)
plt.hist(values, 50)
plt.show()
```

## Distribución exponencial PDF

```python
from scipy.stats import expon
import matplotlib.pyplot as plt

x = np.arange(0, 10, 0.001)
# En x dibujamos el tiempo en ocurrir un hecho. En y la probabilidad de que ocurra,
# por ejemplo: probabilidad de que funcione correctamente un TV
plt.plot(x, expon.pdf(x))
```
